```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam
```

```python
#give seed for random
np.random.seed(0)
n_pts = 500
#create datasets
X, y = datasets.make_circles(n_samples=n_pts, noise=0.1, random_state=123, factor=0.2)
# draw lable 0
plt.scatter(X[y==0, 0], X[y==0, 1])
# draw label 1
plt.scatter(X[y==1, 0], X[y==1, 1])
```

```python
#training ANN model
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(learning_rate=0.01), 'binary_crossentropy', metrics=['accuracy'])
h= model.fit(x=X, y=y, verbose=1, batch_size=50, epochs=500, shuffle="true")
```

```python
#training ANN model
model1 = Sequential()
model1.add(Dense(6, input_shape=(2,), activation='sigmoid'))
model1.add(Dense(1, activation='sigmoid'))
model1.compile(Adam(learning_rate=0.01), 'binary_crossentropy', metrics=['accuracy'])
h1= model1.fit(x=X, y=y, verbose=1, batch_size=50, epochs=500, shuffle="true")
```

```python
# draw graph comparision between 6 neural and 4 neural
plt.plot(h.history['accuracy'])
plt.plot(h1.history['accuracy'])

plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['Model 1 Accuracy', 'Model 2 Accuracy'])

plt.show()
```

```python
# draw graph comparision between 6 neural and 4 neural
plt.plot(h.history['loss'])
plt.plot(h1.history['loss'])

plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['Model 1 Accuracy', 'Model 2 Accuracy'])

plt.show()
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    grid = np.c_[xx.ravel(), yy.ravel()]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])

x = 0
y = 0.75

point = np.array([[x, y]])
predict = model.predict(point)
plt.plot([x], [y], marker='o', markersize=10, color="red")
print("Prediction is: ", predict)
```
